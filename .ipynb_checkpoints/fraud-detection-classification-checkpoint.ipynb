# Introduction

_______________________________________________________________________

Scams target people all over the world. Everyone is vulnerable to scams, so everyone needs information on how to recognize and avoid them. Scammers are smart, and if you don't know what to watch out for, anyone can fall victim to a scam.


Scams work because they look like the real thing and happen when you least expect them. They can include an offer too tempting to be true, a phone call to help fix your computer, or a threat to pay money you don't owe,
Or a warning from your bank or telecom provider about a problem with your account, or even an invitation to "friend" someone online.

Scammers take advantage of new technology, new products or services, and major events to create believable stories that will persuade you to provide your money or personal details.

Scammers even impersonate government officials and make false allegations or use threats of fines, arrest, and deportation to intimidate you into paying. These scammers may get some of your personal information from social media sites to make their demands appear more legitimate.

# LogisticRegression Model


```python
log_cfl = LogisticRegression(C=0.95, random_state=0)

log_cfl.fit(X_train, y_train)

y_predict = log_cfl.predict(X_val)

# evaluate the model
print(classification_report(y_val, y_predict))
print(confusion_matrix(y_val, y_predict))
```

# Random Forest Classifier


```python
rf_cfl = RandomForestClassifier(max_depth=2)
rf_cfl.fit(X_train, y_train)

y_predict = rf_cfl.predict(X_train)

# evaluate the model
print(classification_report(y_train, y_predict))
print(confusion_matrix(y_train, y_predict))
```
